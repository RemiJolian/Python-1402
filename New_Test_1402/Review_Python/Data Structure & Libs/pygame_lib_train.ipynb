import pygame

# Initialize Pygame
pygame.init()

# Set up the display
screen = pygame.display.set_mode((300, 300))
pygame.display.set_caption("Drawing Shapes")
screen.fill((255, 255, 255))

# Set default font size and color
font_size = 1
font_color = (0, 0, 0)

# Initialize the font
font = pygame.font.SysFont("Arial", font_size)

# Main loop
running = True
while running:
    # Get user input
    user_input = input("Enter command: ")

    # Check for valid input
    if user_input.startswith("change size"):
        try:
            new_font_size = int(user_input.split(" ")[2])
            font_size = new_font_size
            font = pygame.font.SysFont("Arial", font_size)
        except ValueError:
            print("Invalid font size")

    elif user_input.startswith("change color"):
        try:
            r, g, b = user_input.split(" ")[2:]
            font_color = (int(r), int(g), int(b))
        except ValueError:
            print("Invalid color")

    elif user_input.startswith("draw line"):
        try:
            x1, y1, x2, y2 = user_input.split(" ")[2:]
            #text = font.render("Line", True, font_color)
            # Draw the text on the screen
            #screen.blit(text, (0, 0))
            pygame.draw.line(screen, font_color, (int(x1), int(y1)), (int(x2), int(y2)), font_size)
        except ValueError:
            print("Invalid line coordinates")

    elif user_input.startswith("draw circle"):
        try:
            x, y, radius = user_input.split(" ")[2:]
            pygame.draw.circle(screen, font_color, (int(x), int(y)), int(radius), font_size)
        except ValueError:
            print("Invalid circle coordinates")

    elif user_input.startswith("draw polygon"):
        try:
            points = []
            coordinates = user_input.split(" ")[2:]
            for i in range(0, len(coordinates), 2):
                x, y = int(coordinates[i]), int(coordinates[i + 1])
                points.append((x, y))
            pygame.draw.polygon(screen, font_color, points, font_size)
        except ValueError:
            print("Invalid polygon coordinates")

    elif user_input == "end drawing":
        # Save the drawing to a file
        pygame.image.save(screen, "draw.png")
        running = False

    # Update the display
    pygame.display.update()

# Quit Pygame
pygame.quit()

import re

import pygame


class Paint:
    pygame.init()
    screen = pygame.display.set_mode((300, 300))
    screen.fill((255, 255, 255))
    pygame.display.update()

    def __init__(self):
        self.size = 1
        self.color = (0, 0, 0)

    def change_size(self, size: int):
        self.size = size

    def change_color(self, color: tuple):
        self.color = color

    def draw_line(self, c1: tuple, c2: tuple):
        pygame.draw.line(self.screen, self.color, c1, c2, self.size)

    def draw_circle(self, center: tuple, radius: int):
        pygame.draw.circle(self.screen, self.color, center, radius, self.size)

    def draw_polygon(self, c: list):
        pygame.draw.polygon(self.screen, self.color, c, self.size)

    def save_image(self):
        pygame.image.save(self.screen, 'draw.png')

    def mange_input(self, req: str):
        nums = list(map(int, re.findall("\d+", req)))
        if "change size" in req:
            self.change_size(nums[0])
        elif "change color" in req:
            self.change_color(tuple(nums))
        elif "draw line" in req:
            self.draw_line(tuple(nums[:2]), tuple(nums[2:]))
        elif "draw circle" in req:
            self.draw_circle(tuple(nums[:2]), nums[2])
        elif "draw polygon" in req:
            coordinates = []
            for i in range(0, len(nums), 2):
                coordinates.append(tuple(nums[i:i+2]))
            self.draw_polygon(coordinates)
        else:
            self.save_image()
            return False
        return True

    def run_paint(self):
        flag = True
        while flag:
            req = input()
            flag = self.mange_input(req)
            pygame.display.update()
            pygame.event.pump()


if __name__ == "__main__":
    paint = Paint()
    paint.run_paint()

import pygame

# مقداردهی اولیه به pygame
pygame.init()

# تعریف ثوابت
WIDTH, HEIGHT = 300, 300
WHITE = (255, 255, 255)
DEFAULT_COLOR = (0, 0, 0)
DEFAULT_SIZE = 1

# ایجاد پنجره نمایش
screen = pygame.display.set_mode((WIDTH, HEIGHT))
screen.fill(WHITE)
pygame.display.set_caption("نقاشی با پایتون")

current_color = DEFAULT_COLOR
current_size = DEFAULT_SIZE

while True:
    command = input().split()

    if command[0] == "change" and command[1] == "color":
        current_color = (int(command[2]), int(command[3]), int(command[4]))

    elif command[0] == "change" and command[1] == "size":
        current_size = int(command[2])

    elif command[0] == "draw" and command[1] == "line":
        pygame.draw.line(screen, current_color, (int(command[2]), int(command[3])), (int(command[4]), int(command[5])), current_size)

    elif command[0] == "draw" and command[1] == "circle":
        pygame.draw.circle(screen, current_color, (int(command[2]), int(command[3])), int(command[4]), current_size)

    elif command[0] == "draw" and command[1] == "polygon":
        points = [(int(command[i]), int(command[i + 1])) for i in range(2, len(command), 2)]
        pygame.draw.polygon(screen, current_color, points, current_size)

    elif command[0] == "end" and command[1] == "drawing":
        pygame.image.save(screen, "draw.png")
        break

    pygame.display.flip()
pygame.quit()
